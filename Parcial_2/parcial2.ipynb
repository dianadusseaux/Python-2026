### **Crear la nueva variable sin datos nulos

```python
datos_sin_nulos = datos.dropna() #Con dropna se eliminan los duplicados
datos_sin_nulos #nueva variable sin datos nulos
```

### **Crear una función que permita filtrar los pokemon de tipo fuego 

```python
def filtro_fuego(Fire): #función para acceder a los pokemones de fuego
    
    return "Fire" in Fire

pokemon_fuego = datos_sin_nulos[datos_sin_nulos]["Type_1"]==["Fire"]

#Mostrar los 20 pokemones de la lista
pokemon_fuego.head(20)
```

```python
def nivel_ataque(int): #Creando función para clasificar los niveles de ataque
    return "att" in int
#función nivel de ataque
```

```python
Promedio = pokemon_fuego.mean() #la función mean nos permite obtener el promedio
Promedio
```

### **Creando grafico barra

```python
plt.bar(pokemon_fuego,nivel_ataque) #con bar estamos generando un grafico de barra
plt.title("Top 10 de Pokemon y su nivel de ataque") 
plt.xlabel("Pokemon")
plt.ylabel("Nivel de ataque")
plt.show()
```
