### Eliminar los datos nulos del dataset
Importante guardar el dataset sin nulos dentro de una variable

```python
datos_sin_nulos = datos.dropna() #Se eliminaron 14 filas
```

```python
datos_sin_nulos
```

### Función para verificar si un género de videojuego esta presente en la lista
En el ejemplo vamos a verificar si se encuentra el género RPG en la lista

```python
# Función que filtra por género RPG
def filtro_rpg(genres):
    return "RPG" in genres

# Aplicar un filtro para consultar y llamar la función filtro_rpg()
datosRPG =datos_sin_nulos[datos_sin_nulos["Genres"].apply(filtro_rpg)] # creando un nuevo dataframe con los videojuegos RPG

# Mostrar el dataset filtrado, incluyendo solo las columnas (Título y Rating)
nombresRPG = datosRPG[["Title", "Rating"]]
nombresRPG.head(20) # Que se muestren los primeros 20 videojuegos RPG
```

### Agregando las etiquetas del dato al gráfico

```python
graficoRPG = nombresRPG.head(10) # Grafico RPG solo tiene los primeros 10 videojuegos RPG

graficoRPG.set_index("Title", inplace=True) # Establece la columna "Title" como indice
graph = graficoRPG["Rating"].plot(kind="bar")

#Agregando etiquetas de datos al gráfico
for i, rating in enumerate(graficoRPG["Rating"]):
    graph.annotate(str(rating), xy=(i, rating), ha="center", va="bottom") # annotate agrega str o escribe un texto sobre un plop
    
plt.show()
```

## Gráfico torta

### Promedio del rating de videojuegos RPG

```python
promedioRPG = datosRPG["Rating"] # promedio de videojuegos RPG
promRPG = promedioRPG.mean() # Función de promedio en numpy
promRPG
```

### Promedio del Rating de videojuegos indie

```python
# Función que filtra por genero indie
def filtro_indie(genres):
    return 'Indie' in genres

# Aplicar un filtro para consultar y llamar la función filtro_indie()
datosIndie = datos_sin_nulos[datos_sin_nulos['Genres'].apply(filtro_indie)] # creando un nuevo dataframe con los videojuegos indie

# Mostrar el dataset filtrado, incluyendo solo las columnas (Titulo y Rating)
nombresIndie = datosIndie[['Title', 'Rating']]
nombresIndie.head(20) # que se muestren los primeros 20 videojuegos indie

```

```python
promedioindie = nombresIndie["Rating"] # promedio de videojuegos indie
promindie = promedioindie.mean() # Función de promedio en numpy
promindie
```

### Crear gráfico circular

```python
dataset_promedios = pd.DataFrame({'Genero': ['RPG', 'Indie'], 'Promedio': [promRPG, promindie]})
```

```python
plt.pie(dataset_promedios['Promedio'], labels=dataset_promedios['Genero'], autopct='%1.1f%%')
plt.title("Comparación de Rating entre RPG e Indie")
plt.show()
```
