```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def procesador_caudal():
    horas = np.arange(1,9) #array =[1,2,3,4,5,6,7,8]
    caudales = [12.5, 14.2, 18.1, 22.8, 25.4, 21.0, 17.5, 15.3]
    
    #creando diccionario
    data ={
        'Hora': horas,
        'Caudal': caudales
        }
    
    #Transformando el diccionario a dataframe
    df = pd.DataFrame(data)
    return df
```

```python
def visualizar_curva(dataframe):
    fig, ax = plt.subplots() #ax -->axes / fig-->figura
    
    ax.plot(dataframe['Hora'], dataframe['Caudal'], marker = 'o', linestyle = '--', color ="blue")
    ax.set_title('Monitoreo de Caudal Rio Gamboa')
    ax.set_xlabel("Tiempo (Horas)")
    ax.set_ylabel("Caudal ($m^3/s)") #ax es un objeto por eso se antepone a las funciones
    
    ax.grid(True)
```

```python
df_hidrologia = procesador_caudal() #Invocando la primera función
visualizar_curva(df_hidrologia) #Invocando segunda funcion con el primer parametro
plt.show()
```
