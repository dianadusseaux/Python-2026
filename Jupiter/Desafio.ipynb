### **Importando librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### **Preparación y limpieza de fechas

```python
df = pd.read_csv("data/Games.csv")
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
```

### **Extracción del año

```python
df['year']= df['Release Date'].dt.year
df['year']
```

### **Agrupación y calculo del promedio

```python
promedio = df.groupby('year')['Rating'].mean().dropna()
print(promedio)
```

```python
print(df.columns)
```

### **Creación de un grafico

```python
plt.figure(figsize=(10,5))
plt.plot(promedio.index, promedio.values, 
         marker= 'o',
         linestyle= '-',
         color= "teal")

plt.grid(True)

plt.title("Evolución del Rating promedio de Videojuegos por año")
plt.xlabel("Año Lanzamiento")
plt.ylabel("Rating Promedio")
plt.show()
```
