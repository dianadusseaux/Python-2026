### **Dimensiones, Vectores y Matrices**

#### **Importando Librería numpy**

````python
import numpy as np
````

#### **Generando un Vector Aleatorio de 12 números enteros (Desde 1 al 10) con Random**

El método `randint()` permite generar números aleatorios enteros. El primer parámetro `1` indica el valor mínimo que puede tomar cada número, y el segundo parámetro, `11`, indica el límite superior, aunque este valor no se incluye; por eso los números generados estarán entre 1 y 10. El tercer parámetro, `12`, define la cantidad de elementos que tendrá el arreglo.

````python
vector = np.random.randint(1, 11, 12) # el tercer parametro indica la cantidad de elementos que tendrá el arreglo
vector
````

#### **Obtención de la Dimensión de un Arreglo**

Con el método `shape()` permite obtener la dimensión del arreglo. En otras palabras, permite saber cuántos elementos tiene y cómo están organizados. 

En este caso, como `vector` es un arreglo unidimensional con 12 valores, `np.shape(vector)` devolverá una tupla con un solo elemento: `(12,)`. El número 12 indica la cantidad de posiciones que contiene el arreglo.

````python
np.shape(vector)
````

#### **Transformar el vector de 12 elementos en una Matriz 4x3**

El método `reshape()` redimensiona una estructura. En este caso toma el arreglo unidimensional `vector` y lo reorganiza en una matriz con 2 filas y 6 columnas. El método `reshape(2, 6)` no cambia los valores que contiene, sino solo su forma, y esto funciona porque el vector tiene 12 elementos, que coinciden con $2 × 6 = 12$.

El resultado se guarda en la variable `matriz`. Es decir, convierte un vector en una matriz sin perder datos, lo que resulta útil para trabajar con datos organizados en filas y columnas.

````python
matriz = vector.reshape(2,6)
matriz
````

#### **Otra variante de Matriz (6x2) generada a partir del vector**

````python
matriz2 = vector.reshape(6,2)
matriz2
````

#### **Volviendo a consultar la dimensión pero de la Matriz**

`matriz` fue creada con `reshape(2, 6)`, entonces `np.shape(matriz)` devolverá (2, 6), lo que significa que la matriz tiene 2 filas y 6 columnas. 

````python
np.shape(matriz)
````

## **Conociendo la Biblioteca Matplotlib**
El paquete ```matplotlib``` permite crear visualizaciones (gráficos) para interpretar resultados y data.

````python
import matplotlib.pyplot as plt
````

#### **Creando un Gráfico de Líneas**

````python
vector2 = np.random.randint(5, 21, 12)
````

`plt.plot(vector, label='Vector')` lo que hace es dibujar una línea que representa los valores del vector, usando sus posiciones como eje horizontal de forma automática. 

Luego el argumento `label='Vector'` asigna un nombre a la serie para que luego pueda aparecer en la leyenda.

````python
plt.plot(vector, label='Vector')
plt.title("Primer Gráfico en MatplotLib") # agrega título al gráfico
plt.xlabel("Indice (Posición)") # agrega el título del eje x
plt.ylabel("Valores") # agrega el título del eje y

plt.grid(True) # para agregar una grilla al gráfico
plt.legend() # mostrar una etiqueta, que se obtiene de la propiedad label del método plot()

plt.show() # permite visualizar gráfico
````

### **Crear dos vectores aleatorios en el mismo gráfico**

````python
vector1 = np.random.randint(1, 11, 12)
vector2 = np.random.randint(5, 21, 12)
````

````python
plt.plot(vector1, marker = 'o', label = "Vector 1")  # la propiedad marker, agrega un marcador en los valores del vector
plt.plot(vector2, marker = 'o', label = "Vector 2")

plt.title("Comparando dos Vectores")
plt.xlabel("Indice (Posición)") 
plt.ylabel("Valores") 

plt.grid(True) 
plt.legend() 

plt.show() 
````
