### **Identificando las columnas con valores nulos

```python
columnas_nulas = datos.isnull().any()
columnas_nulas
```

### **Identificando lasa filas con valores nulos

```python
filas_nulas = datos[datos.isnull().any(axis=1)]
filas_nulas
```

### **Eliminar los datos nulos del dataset

```python
data_sin_nulos = datos.dropna()
data_sin_nulos
```

### **Verificar si un genero del videojuego esta presente en el dataset

```python
def filtro_habilidad(lista_habilidades):
    return "Overgrow" in lista_habilidades

datos_filtrados =data_sin_nulos[data_sin_nulos["Abilities"].apply(filtro_habilidad)]

nombresabilities = datos_filtrados[["Name", "Abilities"]]
nombresabilities.head(20)
```

### **Consultando el nuevo dataset

### **Graficar los primeros 10 pokemon

```python
grafico = nombresabilities.head(10)
grafico
grafico.set_index('Name', inplace=True)
grafico['Abilities'].plot(kind='bar')

```
