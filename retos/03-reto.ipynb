```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def preparar_datos():
    Horas = np.arange(1, 11)
    
    Uso_cpu = [45, 58, 52, 70, 85, 92, 78, 64, 61, 50]
    
    df = pd.DataFrame({
        'Hora':Horas,
        'Uso_Cpu' :Uso_cpu
    })
    return df
```

```python
def graficar_rendimiento(df):
    fig,ax = plt.subplots()
    
    ax.plot(
        df['Hora'],
        df['Uso_Cpu'],
        marker = 'o',
        linestyle = '--',
        color = 'pink'
    )
    
    ax.set_title("Tasa de uso memoria CPU durante 10 horas")
    ax.set_xlabel("Horas de simulación")
    ax.set_ylabel("% Uso CPU")
    ax.grid(True)
    fig.savefig("grafico01",dpi=300)
```

```python
df_rendimiento = preparar_datos()
graficar_rendimiento(df_rendimiento)
plt.show()
```
