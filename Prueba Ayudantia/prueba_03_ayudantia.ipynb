```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lenguaje = ["python", "c++", "java", "JavaScrip", "c#"]
preferencia = [45,20,15,12,8]
print(f"Los lenguajes de programación son: {lenguaje}")
print(f"Las preferencias de los lenguajes de programación son: {preferencia}")
```

```python
plt.bar(lenguaje,preferencia, color=["red","yellow","blue","orange","pink"])
plt.title("Preferencias en los los lenguajes de programación")
plt.xlabel("Lenguaje de programación")
plt.ylabel("Preferencias de lenguajes de programación (%)")
plt.show()
```
